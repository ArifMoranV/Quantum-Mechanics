```julia
# Define the quantum numbers
n = 4
l = 2
m = 1

# Constants
a0 = 1.0
N = 100
phi = range(0, stop=2π, length=N * 5)
th = 0  # You can set this to 0 since it's not being used in your code
r = range(0.001, stop=15, length=N)

# Create 3D grids
RR, PHI, TH = ndgrid(r, phi, th)

# Calculate the spherical harmonic
Yml = sph_harm(m, l, TH, PHI)

# Calculate the normalization constant
coeff = sqrt((2 / (n * a0))^3 * factorial(n - l - 1) / (2 * n * factorial(n + l)))

# Calculate the associated Laguerre polynomial
laguerre_poly = assoc_laguerre(2 * RR / (n * a0), n - l - 1, 2 * l + 1)

# Calculate the radial wavefunction
u = coeff * exp.(-RR / (n * a0)) .* (2 * RR / (n * a0)).^l .* laguerre_poly

# R is equivalent to u/RR in your Python code
R = u ./ RR
```
